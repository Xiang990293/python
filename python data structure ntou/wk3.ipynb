```python
d = {"chicken": 2, "dog": 4, "cat": 4, "spider": 6, "snail": 0}
for animal in d:
    print(animal, d[animal])
for animal, lags in d.items():
    print(f"{animal} have {lags} lags")
```

```python
import numpy as np

x = np.linspace(0,10,50)
sinus = np.sin(x)
cosinus = np.cos(x)
plt.plot(x, sinus, x, cosinus)
plt.show()
```

```python
x = np.linspace(0, 10, 50)
sinus = np.sin(x)
cosinus = np.cos(x)
plt.plot(x, sinus, "k--s",
         x, cosinus, ",c-")
plt.show()
```

```python
plt.title("01181020  sin and cos in [0, 10]")
plt.plot(x, sinus, "k--s",
         x, cosinus, ",c-")
plt.xlabel("radian (rad)")
plt.ylabel("amplitude")
plt.show()
```
